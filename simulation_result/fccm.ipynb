```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load data
scaling_quant_result = pd.read_csv('quantization_inout_en_de3.csv')

# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # Two subplots side by side

# First subplot: OPUS-MT-EN-DE plot
ax1 = axes[0]
baseline_bleu = 41.337328250540224

# Draw baseline and compression ratio reference lines
ax1.axhline(baseline_bleu, color='red', linestyle='--', linewidth=1, alpha=0.8, label='FP32 Baseline')

# Set labels
ax1.set_title('OPUS-MT-EN-DE')
ax1.set_xlabel('Weight Word Length')
ax1.set_ylabel('BLEU Score')

act_wls = [16, 8, 6]
# Use Viridis colormap
viridis = cm.get_cmap('viridis', len(act_wls))  # Get 4 distinct colors from Viridis

for i, act_wl in enumerate(act_wls):
    # Plot BLEU scores for different scaling methods with rounded dots
    ax1.plot(
        scaling_quant_result[
            (scaling_quant_result["Weight Word Length"] != 2) &
            (scaling_quant_result["Activation Word Length"] == act_wl)
        ]["Weight Word Length"],
        scaling_quant_result[
            (scaling_quant_result["Weight Word Length"] != 2) &
            (scaling_quant_result["Activation Word Length"] == act_wl)
        ]["BLEU Score (Range-Based)"],
        linewidth=2,
        alpha=0.8,
        label=f'Activation Word Length {act_wl}',
        color=viridis(i), marker='D', linestyle='-'  # Round dots for data points
    )

# Add legend outside the plot
ax1.legend()  # Legend inside the subplot
ax1.grid(linestyle='--', alpha=0.7)

# Second subplot: Leave it blank
ax2 = axes[1]

baseline_bleu = 31.295117829679363

# Draw baseline and compression ratio reference lines
ax2.axhline(baseline_bleu, color='red', linestyle='--', linewidth=1, alpha=0.8, label='FP32 Baseline')
scaling_quant_result = pd.read_csv('quantization_inout_fr_en3.csv')
# Set labels
ax2.set_title('OPUS-MT-FR-EN')
ax2.set_xlabel('Weight Word Length')
ax2.set_ylabel('BLEU Score')

act_wls = [16, 8, 6]
# Use Viridis colormap
viridis = cm.get_cmap('viridis', len(act_wls))  # Get 4 distinct colors from Viridis

for i, act_wl in enumerate(act_wls):
    # Plot BLEU scores for different scaling methods with rounded dots
    ax2.plot(
        scaling_quant_result[
            (scaling_quant_result["Weight Word Length"] != 2) &
            (scaling_quant_result["Activation Word Length"] == act_wl)
        ]["Weight Word Length"],
        scaling_quant_result[
            (scaling_quant_result["Weight Word Length"] != 2) &
            (scaling_quant_result["Activation Word Length"] == act_wl)
        ]["BLEU Score (Range-Based)"],
        linewidth=2,
        alpha=0.8,
        label=f'Activation Word Length {act_wl}',
        color=viridis(i), marker='D', linestyle='-'  # Round dots for data points
    )

# Add legend outside the plot
ax2.legend()  # Legend inside the subplot
ax2.grid(linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()

# Save the figure to a file
plt.savefig('/Users/justin/Desktop/FCCM/section1/quantisation.png', dpi=500, bbox_inches='tight')  # Save as a PNG file

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load data
scaling_quant_result = pd.read_csv('quantization_weight_only_scaling1.csv')

# Set up the plot
plt.figure(figsize=(10, 5))

baseline_bleu = 43.10655025189425

# Draw baseline and compression ratio reference lines
plt.axhline(baseline_bleu, color='red', linestyle='--', linewidth=2, alpha=0.5, label='FP32 Baseline')
plt.axvline(1, color='gray', linestyle=':', linewidth=4, alpha=0.5)

# Set labels
plt.title('BLEU Score vs Compression Ratio (Range-Based Scaling)')
plt.xlabel('Compression Ratio')
plt.ylabel('BLEU Score')

# Use Viridis colormap
viridis = cm.get_cmap('viridis', 4)  # Get 4 distinct colors from Viridis

# Plot BLEU scores for different scaling methods with rounded dots
plt.plot(
    scaling_quant_result["Compression Ratio"],
    scaling_quant_result["BLEU Score (Range-Based)"],
    linewidth=2,
    alpha=0.5,
    label='Quantisation Baseline',
    color='red', linestyle='-'  # Round dots for data points
)


# Load the SVD data
svd_data1 = pd.read_csv('svd_quantization_results_scaling_en_de1.csv')
svd_data2 = svd_data1


# Define the word lengths to plot for the SVD data
svd_word_lengths = [8, 6, 4, 3]
svd_word_lengths1 = [8, 6, 4, 3]

# Generate a colormap with Viridis for the SVD plots
svd_viridis = cm.get_cmap('viridis', len(svd_word_lengths))

# Create a scatter plot for each word length in the SVD data
for idx, (wl1, wl2) in enumerate(zip(svd_word_lengths,svd_word_lengths1)):
    # Filter the data for the current word length
    
    svd_data1_wl = svd_data1[svd_data1['Word Length'] == wl1]
    svd_data1_wl1 = svd_data2[svd_data2['Word Length'] == wl2]

    # Scatter plot for the current word length
    plt.scatter(
        svd_data1_wl['Compression Ratio'],
        svd_data1_wl["Quant SVD BLEU (Range-Based)"],
        s=170,
        marker='o',
        alpha=0.2,
        color=svd_viridis(idx / len(svd_word_lengths)),
        label=f'Quant SVD\nWord Length {wl1}'
    )

    # Scatter plot for the current word length
    plt.scatter(
        svd_data1_wl1['Compression Ratio'],
        svd_data1_wl1["Iterative Quant SVD BLEU (Range-Based)"],
        s=100,
        marker='o',
        alpha=1,
        color=svd_viridis(idx / len(svd_word_lengths)),
        label=f'Iterative Quant SVD\nWord Length {wl2}'
    )



# Add legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))  # Legend outside plotting area
plt.grid(linestyle='--', alpha=0.7)
plt.xlim([0,15])
plt.ylim([20,45])

# Adjust layout and show the plot
plt.tight_layout(rect=[0, 0, 1, 1])  # Leave space for legend
plt.show()
```
