```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load data
scaling_quant_result = pd.read_csv('quantization_inout_en_de3.csv')

# Set up the plot
plt.figure(figsize=(7, 5))
baseline_bleu = 41.337328250540224

# Font size for labels and ticks
label_font_size = 12
tick_font_size = 12

# Draw baseline and compression ratio reference lines
plt.axhline(baseline_bleu, color='red', linestyle='--', linewidth=2, alpha=0.5, label='FP32 Baseline')
plt.axvline(1, color='gray', linestyle=':', linewidth=4, alpha=0.5)

# Set labels
plt.title('BLEU Score vs Compression Ratio (Activation Word Length 8)', fontsize=label_font_size)
plt.xlabel('Compression Ratio', fontsize=label_font_size)
plt.ylabel('BLEU Score', fontsize=label_font_size)

# Set tick font size
plt.tick_params(axis='both', which='major', labelsize=tick_font_size)

# Use Viridis colormap
viridis = cm.get_cmap('viridis', 4)  # Get 4 distinct colors from Viridis

act_wl = 8

# Plot BLEU scores for different scaling methods with rounded dots
plt.plot(
    scaling_quant_result[(scaling_quant_result["Activation Word Length"] == act_wl)]["Compression Ratio"],
    scaling_quant_result[(scaling_quant_result["Activation Word Length"] == act_wl)]["BLEU Score (Range-Based)"],
    linewidth=2,
    alpha=0.5,
    label='Quantisation Baseline',
    color='red', linestyle='-'  # Round dots for data points
)


# Load the SVD data
svd_data1 = pd.read_csv('svd_quant_inout_en_de_3.csv')
# svd_data1 = pd.concat([svd_data1, pd.read_csv('svd_quant_inout_en_de_4.csv')], ignore_index=True)
# svd_data1 = pd.concat([svd_data1, pd.read_csv('svd_quant_inout_en_de_5.csv')], ignore_index=True)
# svd_data1 = pd.concat([svd_data1, pd.read_csv('svd_quant_inout_en_de_6.csv')], ignore_index=True)
# svd_data1 = pd.concat([svd_data1, pd.read_csv('svd_quant_inout_en_de_7.csv')], ignore_index=True)
# svd_data1 = pd.concat([svd_data1, pd.read_csv('svd_quant_inout_en_de_8.csv')], ignore_index=True)
svd_data2 = svd_data1


# Define the word lengths to plot for the SVD data
svd_word_lengths = [3]
svd_word_lengths1 = [3]

# Generate a colormap with Viridis for the SVD plots
svd_viridis = cm.get_cmap('viridis', len(svd_word_lengths))

# Create a scatter plot for each word length in the SVD data
for idx, (wl1, wl2) in enumerate(zip(svd_word_lengths,svd_word_lengths1)):
    # Filter the data for the current word length
    
    svd_data1_wl = svd_data1[(svd_data1['Activation Word Length'] == act_wl) & (svd_data1['Weight Word Length'] == wl1)]
    svd_data1_wl1 = svd_data2[(svd_data2['Activation Word Length'] == act_wl) & (svd_data2['Weight Word Length'] == wl2)]

    # Scatter plot for the current word length
    plt.scatter(
        svd_data1_wl['Compression Ratio'],
        svd_data1_wl["Quant SVD BLEU (Range-Based)"],
        s=170,
        marker='o',
        alpha=0.2,
        color=svd_viridis(idx / len(svd_word_lengths)),
        label=f'Quant SVD\nWord Length {wl1}'
    )

    # Scatter plot for the current word length
    plt.scatter(
        svd_data1_wl1['Compression Ratio'],
        svd_data1_wl1["Iterative Quant SVD BLEU (Range-Based)"],
        s=100,
        marker='o',
        alpha=1,
        color=svd_viridis(idx / len(svd_word_lengths)),
        label=f'Iterative Quant SVD\nWord Length {wl2}'
    )

optimal = pd.read_csv('optimal_results.csv')
plt.scatter(
    optimal['Compression Ratio'],
    optimal["Optimal BLEU Score"],
    s=100,
    marker='D',
    alpha=1,
    color=svd_viridis(idx / len(svd_word_lengths)),
    label=f'Iterative Quant SVD (LWO)\nWord Length {wl2}'
)


# Add legend outside the plot
plt.legend(loc='lower left',fontsize=12)  # Legend outside plotting area
plt.grid(linestyle='--', alpha=0.7)
plt.xlim([4,15])
plt.ylim([25,43])

# Adjust layout and show the plot
plt.tight_layout(rect=[0, 0, 1, 1])  # Leave space for legend

# Save the figure to a file
plt.savefig('/Users/justin/Desktop/FCCM/section6/svd_quant.png', dpi=500, bbox_inches='tight')  # Save as a PNG file

plt.show()
```
